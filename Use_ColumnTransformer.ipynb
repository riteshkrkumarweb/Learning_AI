```
X = df[['Hours_Studied',	'Attendance',	'Sleep_Hours',	'Previous_Scores',	'Family_Income',	'School_Type',	'Gender']]
y = df['Exam_Score']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```
from sklearn.compose import ColumnTransformer
# import the required libraries

import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```
from scipy import sparse
transformer = ColumnTransformer(transformers=[
    ('tnf0',SimpleImputer(),['Attendance']),
    ('tnf3',OrdinalEncoder(categories=[['Low','Medium','High']]),['Family_Income']),
    ('tnf6',OneHotEncoder(sparse_output=False,drop='first'),['Gender','School_Type']),
],remainder='passthrough')
```

```
transformer.fit_transform(X_train).shape
```

```
transformer.transform(X_test).shape
```
