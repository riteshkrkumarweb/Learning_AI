```python
# 📊 Bagging (Bootstrap Aggregating)

# Bagging is an Ensemble Learning technique

# 🎯 Goal:
# Improve accuracy and reduce overfitting

# 🧠 Idea:
# Train multiple models on different random subsets
# of the dataset and combine their predictions

# 📦 Bootstrap:
# Random sampling WITH replacement

# Means:
# Same data point can appear multiple times
# in one sampled dataset

# 📊 Aggregating:
# Combine predictions from all models

# For Classification:
# → Majority Voting 🗳️

# For Regression:
# → Average Prediction 📈


# 🔁 Working of Bagging:

# Step 1️⃣
# Create multiple random datasets from original data

# Step 2️⃣
# Train separate models on each dataset

# Step 3️⃣
# Collect predictions from all models

# Step 4️⃣
# Combine predictions for final output


# 📦 Commonly used with:
# 🌳 Decision Trees

# Example:
# Random Forest = Bagging + Decision Trees


# 🎯 Advantages:
# ✅ Reduces overfitting
# ✅ Improves stability
# ✅ Better accuracy
# ✅ Works well on high variance models

# ❌ Disadvantages:
# ❌ Slower
# ❌ More memory usage
# ❌ Less interpretable


# 📊 Simple Diagram:

#                📥 Original Dataset
#                         │
#       ┌─────────────────┼─────────────────┐
#       │                 │                 │
#   📦 Sample1        📦 Sample2        📦 Sample3
#       │                 │                 │
#    🌳 Tree1          🌳 Tree2          🌳 Tree3
#       │                 │                 │
#       └─────────────────┼─────────────────┘
#                         │
#               🗳️ Voting / Averaging
#                         │
#                  ✅ Final Prediction
```

# **Type of Bagging**

```python
# =========================================================
#              TYPES OF ENSEMBLE LEARNING
# =========================================================


# 1️⃣ BAGGING
# ---------------------------------------------------------
#
#          Random Samples from Dataset
#
#        ┌────────┬────────┬────────┐
#        ▼        ▼        ▼
#      Tree1    Tree2    Tree3
#
#        └────────┼────────┘
#                 ▼
#          Voting / Average
#                 ▼
#          Final Prediction
#
# Goal:
# - Reduce Variance
# - Reduce Overfitting
#
# Example:
# - Random Forest
#
# =========================================================


# 2️⃣ BOOSTING
# ---------------------------------------------------------
#
#          Model 1
#             ▼
#      Correct Mistakes
#             ▼
#          Model 2
#             ▼
#      Correct Mistakes
#             ▼
#          Model 3
#             ▼
#      Final Prediction
#
# Goal:
# - Reduce Bias
#
# Examples:
# - AdaBoost
# - XGBoost
#
# =========================================================


# 3️⃣ VOTING
# ---------------------------------------------------------
#
#        KNN  ────┐
#        Tree ────┼──► Voting ─► Final Prediction
#        SVM  ────┘
#
# Goal:
# - Combine Different Algorithms
#
# =========================================================


# 4️⃣ STACKING
# ---------------------------------------------------------
#
#        KNN  ────┐
#        Tree ────┼──► Meta Model ─► Final Prediction
#        SVM  ────┘
#
# Goal:
# - Learn Best Combination of Models
#
# =========================================================
```

# Simple Bagging with DicisionTree

# OBB score
OOB Score (Out-Of-Bag Score) is the accuracy or performance of a bagging model measured using the rows/samples that were not selected during bootstrap training of a particular tree or estimator. you can say that like the X_test

⚠️ OOB works with:
✅ rows/samples
❌ not columns/features

🔥 Used mainly in:
Bagging
Random Forest

```python
bag_e = BaggingClassifier(
estimator=DecisionTreeClassifier(), #Creating Decision Tree as base estimator
n_estimators=500, #Creating 500 Decision Tree models
max_samples=0.25, #Taking 25% random rows for each model
bootstrap=True, #Enabling sampling with replacement
oob_score=True, #Enabling Out Of Bag score calculation
max_features=0.5,   
random_state=42 #Fixing random behavior for same results
)
bag_e.fit(X_train,y_train)
y_pred_e = bag_e.predict(X_test)
```

```python
print("Random Patches classifier", accuracy_score(y_test,y_pred_e))
print(('OBB score is ',bag_e.oob_score_))
```
