```python
# 📊 Naive Bayes Algorithm

# "Naive" → Assumes all features are independent
# "Bayes" → Based on Bayes Theorem (probability concept)

# 🎯 Used for Classification problems
# Example:
# Spam / Not Spam 📧
# Positive / Negative sentiment 😊😠

# 🧠 Idea:
# It calculates probability of each class
# and picks the class with highest probability

# Formula:
# P(C|X) = [P(C) * P(X|C)] / P(X)

# Meaning:
# P(C|X) → Probability of class C given data X
# P(C)   → Prior probability of class
# P(X|C) → Likelihood (data given class)
# P(X)   → Evidence (constant)

# Since P(X) is same for all classes,
# we compare only:
# P(C) * P(X|C)

# 🔑 Naive assumption:
# Features are independent from each other

# Example:
# Email → "free", "offer", "win"
# Naive Bayes assumes these words are independent

# 📦 Types of Naive Bayes:

# 1. Gaussian Naive Bayes
# → For continuous data (numbers)

# 2. Multinomial Naive Bayes
# → For discrete counts (text data, word frequency)

# 3. Bernoulli Naive Bayes
# → For binary features (0/1, yes/no)

# ⚡ Advantages:
# Fast
# Works well on text data
# Simple to implement

# ⚠️ Limitation:
# Assumption of independence is not always true
```

```python
df = pd.read_csv('/kaggle/input/datasets/fedesoriano/company-bankruptcy-prediction/data.csv')
```

```python
X = df.iloc[:,1:]
X
```

```python
y = df.iloc[:,0]
y
```

```python
from sklearn.model_selection import  train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
df.shape
```

```python
X_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
transformer1 = GaussianNB()
```

```python
pipe = Pipeline([
    ('trf1',transformer1)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_predic = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_predic)
```

```python
confusion_matrix(y_test,y_predic)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predic))
```
