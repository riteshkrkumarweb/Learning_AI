```python
#Mini_Batch_Gredient_Descent
import random
from sklearn.linear_model import SGDRegressor   # Import the SGDRegressor model (uses gradient descent for regression)

sgd = SGDRegressor(learning_rate='constant', eta0=0.1)
# Create the model
# learning_rate='constant' → learning rate will not change during training
# eta0 = 0.1 → step size used when updating weights

batch_size = 35
# Number of rows the model will train on at one time (mini-batch size)

for i in range(100):
    # Run the training loop 100 times (100 mini-batch updates)

    idx = random.sample(range(X_train_scaled.shape[0]), batch_size)
    # X_train.shape[0] → total number of rows in the training dataset
    # range(X_train_scaled.shape[0]) → creates numbers from 0 to (total_rows - 1)
    # random.sample(..., batch_size) → randomly select 35 row numbers
    # idx now contains random indexes like [12, 455, 78, 901, ...]

    sgd.partial_fit(X_train_scaled[idx], y_train.iloc[idx])
    # X_train_scaled.iloc[idx] → select those random rows from X_train
    # y_train_scaled.iloc[idx] → select the corresponding target values
    # partial_fit() → train the model using only these 35 rows
    # this updates the model weights slightly

y_pred = sgd.predict(X_test_scaled)
# Use the trained model to predict outputs for the test dataset

score = r2_score(y_test, y_pred)
# Compare predicted values with real values using R² score
# R² tells how good the model predictions are

print(f'The Score is {score}')
# Print the model performance

#Mine is too Worse 😂
```

```python
#This is not working beacuse it is only for the Single feature
class GDRegressor:

    def __init__(self, learning_rate, epochs):
        # Initial values of slope (m) and intercept (b)
        # These are just starting guesses
        self.m = 100
        self.b = -120

        # learning rate → step size for gradient descent
        self.lr = learning_rate

        # number of iterations we want to train
        self.epochs = epochs

    def fit(self, X, y):
        # Training function

        for i in range(self.epochs):

            # X.ravel() converts X from column shape to 1D array
            # Example: [[2],[3],[4]] → [2,3,4]

            # Calculate slope of loss function with respect to b
            loss_slope_b = -2 * np.sum(y - self.m * X.values.ravel() - self.b)

            # Calculate slope of loss function with respect to m
            loss_slope_m = -2 * np.sum((y - self.m * X.values.ravel() - self.b) * X.ravel())

            # Update intercept using gradient descent formula
            self.b = self.b - (self.lr * loss_slope_b)

            # Update slope using gradient descent formula
            self.m = self.m - (self.lr * loss_slope_m)

        # After training print the learned parameters
        print(self.m, self.b)

    def predict(self, X):
        # Linear regression prediction formula
        return self.m * X + self.b


# Create model object
gd = GDRegressor(0.001, 50)

# Train model
gd.fit(X_train, y_train)


# Make predictions
y_pred = gd.predict(X_test)

# Evaluate model
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
